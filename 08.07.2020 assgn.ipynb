```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("panda.jpg")
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
```

```python
plt.imshow(img[:,:,0],cmap='gray')
```
