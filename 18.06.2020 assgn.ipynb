### 10 Encode the categorical columns using different kinds of encoding techniques from pandas and scikit-learn


```python
merge_data=pd.get_dummies(data,columns=['Restaurant','Reviewer','Review','Rating','Metadata','Time','Pictures'])
merge_data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # creating an object
merge_data=sc.fit_transform(merge_data)
merge_data
```

```python
merge_data=pd.DataFrame(merge_data)
merge_data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

c_t=ColumnTransformer(transformers=[('Encoder',OneHotEncoder(),['Restaurant'])])

data1=c_t.fit_transform(data)
data1=pd.DataFrame(data1)
data1=data1.iloc[:,0:2]
data1.column=['Rating']
data1=pd.concat([data,data1],axis=1)
data1=data1.drop('Restaurant',axis=1)
data1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1.Review=le.fit_transform(data1.Review)
data1 
```
